# Image Reading - Generally Used for Gray Image

```python
import numpy as np
from PIL import Image 
img = Image.open('./Image/lena.png') 
img_array=np.array(img,dtype=np.uint8)
print(img_array)
print(img_array.shape)
# Output Images 
img.show() 
img_array = img_array.astype(np.uint8)
img=Image.fromarray(img_array)

img.save("Image/lenaDup.png")
```

# Image Reading - Used for Realtime Images

```python
import numpy as np
import cv2 
img = cv2.imread('./Image/lena.png') 
img_array=np.array(img,dtype=np.float32)
print(img_array)
print(img_array.shape)

# Output img with window name as 'image' 
cv2.imshow('image', img)  
  
# Maintain output window utill 
# user presses a key 
cv2.waitKey(0)         
  
# Destroying present windows on screen 
cv2.destroyAllWindows() 
```

# Negation of the Image
- Subtracting pixel values from maximum pixel value

```python
import numpy as np
from PIL import Image

img = Image.open("./Image/boat.png")

img_array=np.array(img,dtype=np.float32)

img_array[:,:] =255-img_array[:,:]

img_array = img_array.astype(np.uint8)

img=Image.fromarray(img_array)

img.save("./Image/boatNegative.png")
```
